```python
from datasets import load_dataset
data=load_dataset('FahadIqbal5188/floorplan-SDXL' , split='train')
data
```

```python
data[0]['image']
```

```python
import os
from tqdm import tqdm

if not os.path.exists('images'):
    os.makedirs('images')

for i in tqdm(range(1628)):
    file_path = f'images/image_{i}.png'
    if not os.path.exists(file_path):  
        data[i]['image'].save(file_path)
```

```python
type(data[0]['image'])
```

```python
import pandas as pd

rows = []
for i in range(1628):
    text_value = data[i]['text']
    if text_value:  
        rows.append({"id": i, "description": text_value})

df = pd.DataFrame(rows)
df.dropna(inplace=True)
df.drop_duplicates(subset="description", inplace=True)

df["image_path"] = [f"images/image_{i}.png" for i in df["id"]]

df.to_csv("cleaned_dataset_part1.csv", index=False)
print(" cleaned_dataset_part1.csv !")
```
